```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'],
                   'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   'Airline': ['KLM(!)', ' (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
flight = df.FlightNumber      #SHALLOW copy
number = flight[0]
for i in range(len(flight)):
    flight[i] = number
    number+=10
df.FlightNumber.astype('int64')
df
```

```python
data = pd.DataFrame(columns=['From','To'])
```

```python
for i in range(len(df.From_To)):
    l = df.From_To[i].split('_')
    
    data = data.append({'From':l[0], 'To':l[1]} ,ignore_index=True)
    #data.From, data.To[i+1] = l[0],l[1]
data    
```

```python
data.From = pd.Series([x.capitalize() for x in list(data.From)])
data.To = pd.Series([x.capitalize() for x in list(data.To)])
data
```

```python
df.drop('From_To', axis=1,inplace= True)
df
```

```python
df = df.join(data, how='inner')
```

```python
def get_values(listt):
    a = []
    for i in range(3):
        try:
            a.append(listt[i])
        except:
            a.append(np.nan)
    return a
delay_1 = []
delay_2 = []
delay_3 = []
for i in range(len(df.RecentDelays)):
    a = get_values(df.RecentDelays[i])
    delay_1.append(a[0])
    delay_2.append(a[1])
    delay_3.append(a[2])   
delays = pd.DataFrame({'delay_1':delay_1,'delay_2':delay_2, 'delay_3':delay_3})
print(delays)
```

```python
df.drop('RecentDelays',axis=1,inplace= True)
df
```

```python
df=df.join(delays,how='inner')
df
```
